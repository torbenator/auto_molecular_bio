```python
%pylab inline
import auto_mol_bio as amb
import random
import re
import matplotlib.pyplot as plt
from collections import OrderedDict
```

```python
T7_feature_dict = {
    'name':'T7',
    'sequence':amb.data.common_promoters['T7'],
    'placement':30
    }

SP6_feature_dict = {
    'name':'SP6',
    'sequence':amb.data.common_promoters['SP6'],
    'placement':600
    }


#print gene_sequence[feature_dict['placement']:feature_dict['placement']+10]
gene_sequence = amb.data.generate_gene_sequence(g_length=1000,
                                                features=[T7_feature_dict,SP6_feature_dict])
promoter_ind_dicts = amb.analysis.find_all_perfect_matches(gene_sequence,[amb.data.common_promoters['T7'],amb.data.common_promoters['SP6']],custom_labels=['T7 Promoter','SP6 Promoter'], meta_label='Promoters')
alanine_inds = amb.analysis.find_all_perfect_matches(gene_sequence,['GCT', 'GCC', 'GCA', 'GCG'],meta_label="Sequences that code for Alanine")
glycine_inds = amb.analysis.find_all_perfect_matches(gene_sequence,['GGT', 'GGC', 'GGA', 'GGG'],meta_label="Sequences that code for Glycine")
```

```python


amb.plotting.plot_genome_features(gene_sequence,feature_dicts=[promoter_ind_dicts,alanine_inds,glycine_inds])
```
